```python
import os

# Caminho para o arquivo de entrada
input_file = './dados_extraidos/v_clientes_CodEmpresa_92577.csv'

# Caminho para a pasta de saída
output_dir = './output'
os.makedirs(output_dir, exist_ok=True)
output_file = os.path.join(output_dir, 'clientes.xlsx')

# Carregar os dados de 'v_clientes_CodEmpresa_92577.csv'
clientes = pd.read_csv(input_file, delimiter=';', encoding='latin1')

# Criar a nova tabela CLIENTES com as colunas transformadas
clientes_novo = pd.DataFrame({
    'NOME': clientes['razao_social'],
    'CPF CNPJ': clientes['cpf_cnpj'],
    'RG': clientes['rg'],
    'NACIONALIDADE': clientes['nacionalidade'],
    'NASCIMENTO': pd.to_datetime(clientes['nascimento'], errors='coerce', dayfirst=True).dt.strftime('%d/%m/%Y'),
    'ESTADO CIVIL': clientes['estado_civil'],
    'PROFISSÃO': clientes['profissao'],
    'SEXO': '',
    'CELULAR': clientes['telefone2'],
    'TELEFONE': clientes['telefone1'],
    'EMAIL': clientes['email1'],
    'PAIS': '',
    'ESTADO': clientes['uf'],
    'CIDADE': clientes['cidade'],
    'BAIRRO': clientes['bairro'],
    'ENDEREÇO': clientes['logradouro'],
    'CEP': clientes['cep'],
    'PIS PASEP': clientes['pis'],
    'CTPS': '',
    'CID': '',
    'NOME DA MÃE': clientes['nome_mae'],
    'ORIGEM DO CLIENTE': '',
    'ANOTAÇÕES GERAIS': ''
})

# Salvar o arquivo em formato Excel
clientes_novo.to_excel(output_file, index=False, engine='openpyxl')

print(f"Arquivo CLIENTES salvo com sucesso em: {output_file}")

```
